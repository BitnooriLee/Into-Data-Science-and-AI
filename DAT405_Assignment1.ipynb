```python
# Import Data set
from google.colab import drive
drive.mount('/content/gdrive')

root = '/content/gdrive/My Drive/DAT405/'

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings; warnings.filterwarnings(action='once')

from sklearn.model_selection import train_test_split
data = pd.read_csv(root+"life-expectancy-vs-gdp-per-capita.csv")
```

```python
# Prepare Data 
data = data[data['Life expectancy'].notna()]
data = data[data['GDP per capita'].notna()]

# Select 10 countries has biggest population year of 2018 
biggest_population_2018 = data.loc[data['Year'] == 2018]
biggest_population_2018 = biggest_population_2018.sort_values(by=['Population (historical estimates)'],ascending=False)
biggest_population_2018 = biggest_population_2018.head(10)
data_selected_country = data.loc[data['Entity'].isin(biggest_population_2018['Entity'])]
```

```python
# Create as many colors as there are unique Entity(Countries)
countries = np.unique(data_selected_country['Entity'])
colors = [plt.cm.tab10(i/float(len(countries)-1)) for i in range(len(countries))]

# Draw Plot for Each Category
plt.figure(figsize=(16, 10), dpi= 80, facecolor='w', edgecolor='k')

for i, country in enumerate(countries):
    plt.scatter('GDP per capita', 'Life expectancy', 
                data = data_selected_country.loc[data_selected_country['Entity']==country, :], 
                s=20, c=colors[i], label=str(country))

# Decorations
plt.gca().set(xlabel='GDP per capita (US dollar)', ylabel='Life expectancy at Birth (Years)')
plt.xticks(fontsize=12); plt.yticks(fontsize=12)
plt.title("GDP per capita vs life expectancy(2018)", fontsize=22)
plt.legend(fontsize=12)    
plt.show()  
```

When GDP per capita is relatively low, The effect of life expectancy improvement due to GDP growth is very large. However, it can be seen that the effect of increasing life expectancy becomes insignificant after GDP per capita reaches above a certain level. The result can be explained that when a country's GDP reaches a certain level, it reflects the guarantee of a basic standard of living such as food, clean water. After reaching a certain level, when all the basic parts are satisfied, other ancillary elements can affect the life expectancy such as healthcare services, leisure lifestyle.
